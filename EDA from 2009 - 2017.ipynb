```python
# import necessory libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# read csv matches file data
df =pd.read_csv('matches.csv')
```

```python
# no of first five
df.head()
```

```python
# size of fle
df.shape
```

```python
# d type of each feature
df.info()
```

```python
#records players of player of matchers 
df['player_of_match'].value_counts()
```

```python
# top 10 player of match
df['player_of_match'].value_counts()[:10]
```

```python
# top 5 player of match
df['player_of_match'].value_counts()[:5]
```

```python
list(df['player_of_match'].value_counts()[:5].keys())
```

```python
# plotting figure of top five player of match
plt.figure(figsize=(8,5))
plt.bar(list(df['player_of_match'].value_counts()[:5].keys()),list(df['player_of_match'].value_counts()[:5]),color='red')
```

```python
# record of result
df['result'].value_counts()
```

```python
# record of the team who won the toss
df['toss_winner'].value_counts()
```

```python
# records of teams who choosen batting first
batting_first = df[df['win_by_runs']!=0]
```

```python
# plot the groph of the teams who chose batting first
plt.figure(figsize=(8,5))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.show()
```

```python
df.head()
```

```python
# records of team got win after choosing bat first
batting_first['winner'].value_counts()
```

```python
list(batting_first['winner'].value_counts()[:3].keys())
```

```python
#pplotting the graph of team got win after choosing bat first
plt.figure(figsize=(8,5))
plt.bar(list(batting_first['winner'].value_counts()[:3].keys()),list(batting_first['winner'].value_counts()[:3]),color='maroon')
```

```python
# ploting piechart of  team got win after choosing bat first
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()), autopct='%1.0f%%')
plt.show()
```

```python
# records of the who wn by wickets
batting_second = df[df['win_by_wickets']!=0]
```

```python
# top five records of the who wn by wickets
batting_second.head()
```

```python
# plot the graph which won by wickets
plt.figure(figsize=(8,5))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.title('Distribution of wickets')
plt.xlabel('Wickets')
plt.show()
```

```python
# team who do batting second and won the match
batting_second['winner'].value_counts()
```

```python
# plot the graph team who do batting second and won the match
plt.figure(figsize=(7,5))
plt.bar(list(batting_second['winner'].value_counts()[:3].keys()),list(batting_second['winner'].value_counts()[:3]),color='maroon')
plt.show()
```

```python
# plot the piechart team who do batting second and won the match
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()), autopct='%1.0f%%')
plt.show()
```
